```python

from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

# Airlines
spark.createDataFrame([], schema=schemas.airlines_schema) \
    .write.format("delta").mode("overwrite") \
    .saveAsTable("flight_sales_bronze.bronze_airlines_raw")

# Airports
spark.createDataFrame([], schema=schemas.airports_schema) \
    .write.format("delta").mode("overwrite") \
    .saveAsTable("flight_sales_bronze.bronze_airports_raw")

# Flights
spark.createDataFrame([], schema=schemas.flights_schema) \
    .write.format("delta").mode("overwrite") \
    .saveAsTable("flight_sales_bronze.bronze_flights_raw")
```
